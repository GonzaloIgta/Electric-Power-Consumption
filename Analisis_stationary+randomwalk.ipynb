```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("powerconsumption.csv")
df.head()
```

```python
ADF_resultzone1 = adfuller(df["PowerConsumption_Zone1"])
ADF_resultzone2 = adfuller(df["PowerConsumption_Zone2"])
ADF_resultzone3 = adfuller(df["PowerConsumption_Zone3"])

print("-------------- ZONE 1 --------------")
print(f'ADF Statistic Zone 1: {ADF_resultzone1[0]}')
print(f'p-value Zone 1: {ADF_resultzone1[1]}')

print("-------------- ZONE 2 --------------")
print(f'ADF Statistic Zone 2: {ADF_resultzone2[0]}')
print(f'p-value Zone 2: {ADF_resultzone2[1]}')

print("-------------- ZONE 3 --------------")
print(f'ADF Statistic Zone 3: {ADF_resultzone3[0]}')
print(f'p-value Zone 3: {ADF_resultzone3[1]}')


# ES ESTACIONARIA (p value < 0.05)
```

```python
plot_acf(df["PowerConsumption_Zone1"], lags=20, title= "Auto Correlation Zone 1")
plot_acf(df["PowerConsumption_Zone2"], lags=20, title= "Auto Correlation Zone 2")
plot_acf(df["PowerConsumption_Zone3"], lags=20, title= "Auto Correlation Zone 3")
```

```python
plot_pacf(df["PowerConsumption_Zone1"], lags=20, title= "Partial Auto Correlation Zone 1")
plot_pacf(df["PowerConsumption_Zone2"], lags=20, title= "Partial Auto Correlation Zone 2")
plot_pacf(df["PowerConsumption_Zone3"], lags=20, title= "Partial Auto Correlation Zone 3")
```
