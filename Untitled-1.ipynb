```python
"""
SARIMAX Power Consumption Forecasting - Tetuán, Morocco
Dataset: powerconsumption.csv (52,416 obs, 10-min intervals)
Prediction horizon: 24 hours = 144 observations
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings("ignore")

from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_absolute_error, mean_squared_error


# ─────────────────────────────────────────────
# 1. LOAD AND PREPROCESS DATA
# ─────────────────────────────────────────────

df = pd.read_csv("powerconsumption.csv")

df["Datetime"] = pd.to_datetime(df["Datetime"])
df = df.sort_values("Datetime").reset_index(drop=True)
df = df.set_index("Datetime")

# Resample to ensure regular 10-min frequency (fill gaps if any)
df = df.resample("10min").interpolate(method="linear")

print("Dataset shape:", df.shape)
print("Date range:", df.index.min(), "→", df.index.max())
print("\nMissing values:\n", df.isnull().sum())
print("\nFirst rows:\n", df.head())


# ─────────────────────────────────────────────
# 2. EXPLORATORY ANALYSIS
# ─────────────────────────────────────────────

ZONES = ["PowerConsumption_Zone1", "PowerConsumption_Zone2", "PowerConsumption_Zone3"]
ZONE_LABELS = ["Zone 1 (Quads)", "Zone 2 (Smir)", "Zone 3 (Boussafou)"]
COLORS = ["#1f77b4", "#ff7f0e", "#2ca02c"]

fig, axes = plt.subplots(3, 1, figsize=(16, 10), sharex=True)
fig.suptitle("Power Consumption – Tetuán (Full series)", fontsize=14, fontweight="bold")
for ax, zone, label, color in zip(axes, ZONES, ZONE_LABELS, COLORS):
    ax.plot(df.index, df[zone], color=color, lw=0.5, alpha=0.8)
    ax.set_ylabel(label, fontsize=10)
    ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("01_full_series.png", dpi=150, bbox_inches="tight")
plt.close()
print("\n[Saved] 01_full_series.png")


# Zoom into 1 week to see daily patterns
sample = df.loc["2017-01-01":"2017-01-07"]
fig, axes = plt.subplots(3, 1, figsize=(16, 8), sharex=True)
fig.suptitle("Power Consumption – First week (10-min resolution)", fontsize=13)
for ax, zone, label, color in zip(axes, ZONES, ZONE_LABELS, COLORS):
    ax.plot(sample.index, sample[zone], color=color, lw=1.2)
    ax.set_ylabel(label, fontsize=9)
    ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("02_week_sample.png", dpi=150, bbox_inches="tight")
plt.close()
print("[Saved] 02_week_sample.png")


# ─────────────────────────────────────────────
# 3. STATIONARITY TESTS (ADF)
# ─────────────────────────────────────────────

print("\n" + "="*55)
print("AUGMENTED DICKEY-FULLER TEST (stationarity)")
print("="*55)
for zone, label in zip(ZONES, ZONE_LABELS):
    result = adfuller(df[zone].dropna(), maxlag=144, autolag="AIC")
    print(f"\n{label}")
    print(f"  ADF statistic : {result[0]:.4f}")
    print(f"  p-value       : {result[1]:.6f}")
    print(f"  Lags used     : {result[2]}")
    verdict = "Stationary ✓" if result[1] < 0.05 else "NON-stationary ✗"
    print(f"  Conclusion    : {verdict}")


# ─────────────────────────────────────────────
# 4. ACF / PACF PLOTS (identify ARIMA orders)
# ─────────────────────────────────────────────

fig, axes = plt.subplots(3, 2, figsize=(16, 12))
fig.suptitle("ACF and PACF – first 150 lags (~25 hours)", fontsize=13)
for row, (zone, label) in enumerate(zip(ZONES, ZONE_LABELS)):
    series = df[zone].dropna()
    plot_acf(series, lags=150, ax=axes[row, 0], title=f"ACF – {label}", alpha=0.05)
    plot_pacf(series, lags=150, ax=axes[row, 1], title=f"PACF – {label}", method="ywm", alpha=0.05)
plt.tight_layout()
plt.savefig("03_acf_pacf.png", dpi=150, bbox_inches="tight")
plt.close()
print("[Saved] 03_acf_pacf.png")


# ─────────────────────────────────────────────
# 5. TRAIN / TEST SPLIT
#    Test = last 144 obs = 24 hours
# ─────────────────────────────────────────────

HORIZON = 144  # 24 h × 6 obs/h

train = df.iloc[-HORIZON - 90*144 : -HORIZON]
test  = df.iloc[-HORIZON:]

print(f"\nTrain size : {len(train)} obs  ({train.index[0]} → {train.index[-1]})")
print(f"Test size  : {len(test)}  obs  ({test.index[0]} → {test.index[-1]})")


# ─────────────────────────────────────────────
# 6. SARIMAX MODEL
#
#  ARIMA order   (p, d, q) = (2, 0, 0)  ← AR(2) as specified
#  Seasonal order(P, D, Q, m)            ← m=144 (daily cycle)
#  Exogenous features (optional):
#    Temperature, Humidity, WindSpeed, GeneralDiffuseFlows, DiffuseFlows
#
#  NOTE: Seasonal period m=144 (daily) is computationally expensive.
#        A daily seasonal difference (D=1) is used to capture the pattern
#        without fitting a full seasonal AR/MA.
#        Adjust (P,D,Q) based on ACF/PACF diagnostic results.
# ─────────────────────────────────────────────

EXOG_COLS = ["Temperature", "Humidity", "WindSpeed", "GeneralDiffuseFlows", "DiffuseFlows"]

# SARIMAX orders
ORDER         = (2, 0, 0)     # Non-seasonal: AR(2), no differencing, no MA
SEASONAL_ORDER = (1, 1, 1, 144)  # Seasonal: daily cycle (144 steps = 24h)

results_dict = {}

for zone, label, color in zip(ZONES, ZONE_LABELS, COLORS):
    print(f"\n{'='*55}")
    print(f"Fitting SARIMAX for {label} ...")
    print(f"  Order         : {ORDER}")
    print(f"  Seasonal order: {SEASONAL_ORDER}")
    print(f"  Exogenous vars: {EXOG_COLS}")

    y_train = train[zone]
    y_test  = test[zone]
    X_train = train[EXOG_COLS]
    X_test  = test[EXOG_COLS]

    model = SARIMAX(
        endog=y_train,
        exog=X_train,
        order=ORDER,
        seasonal_order=SEASONAL_ORDER,
        enforce_stationarity=False,
        enforce_invertibility=False,
        trend="c"
    )

    fitted = model.fit(disp=False, maxiter=200, method="lbfgs")
    print(fitted.summary().tables[0])

    # Forecast
    forecast = fitted.forecast(steps=HORIZON, exog=X_test)
    forecast.index = y_test.index

    # Metrics
    mae  = mean_absolute_error(y_test, forecast)
    rmse = np.sqrt(mean_squared_error(y_test, forecast))
    mape = np.mean(np.abs((y_test - forecast) / y_test)) * 100

    print(f"\n  MAE  : {mae:,.2f}")
    print(f"  RMSE : {rmse:,.2f}")
    print(f"  MAPE : {mape:.2f}%")

    results_dict[zone] = {
        "fitted"   : fitted,
        "forecast" : forecast,
        "y_test"   : y_test,
        "mae"      : mae,
        "rmse"     : rmse,
        "mape"     : mape,
        "label"    : label,
        "color"    : color
    }


# ─────────────────────────────────────────────
# 7. FORECAST PLOTS
# ─────────────────────────────────────────────

# Context: last 3 days + forecast
CONTEXT_OBS = 3 * 144  # 3 days

fig, axes = plt.subplots(3, 1, figsize=(16, 12), sharex=False)
fig.suptitle("SARIMAX 24-hour forecast vs. actual values", fontsize=14, fontweight="bold")

for ax, zone in zip(axes, ZONES):
    r = results_dict[zone]
    context = df[zone].iloc[-(CONTEXT_OBS + HORIZON):-HORIZON]

    ax.plot(context.index, context.values, color=r["color"], lw=1, alpha=0.6, label="Historical (3 days)")
    ax.plot(r["y_test"].index, r["y_test"].values, color="black",   lw=1.8, label="Actual (24 h)")
    ax.plot(r["forecast"].index, r["forecast"].values, color="red", lw=1.8, ls="--", label="Forecast (24 h)")

    # Confidence interval (in-sample approximate)
    std_resid = fitted.resid.std()
    ax.fill_between(
        r["forecast"].index,
        r["forecast"] - 1.96 * std_resid,
        r["forecast"] + 1.96 * std_resid,
        color="red", alpha=0.12, label="95% CI"
    )

    ax.set_title(
        f"{r['label']}  |  MAE={r['mae']:,.0f}  RMSE={r['rmse']:,.0f}  MAPE={r['mape']:.1f}%",
        fontsize=10
    )
    ax.set_ylabel("Power consumption (kW)")
    ax.legend(loc="upper left", fontsize=8)
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%d %b\n%H:%M"))
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("04_forecast.png", dpi=150, bbox_inches="tight")
plt.close()
print("\n[Saved] 04_forecast.png")


# ─────────────────────────────────────────────
# 8. RESIDUAL DIAGNOSTICS
# ─────────────────────────────────────────────

fig, axes = plt.subplots(3, 2, figsize=(16, 12))
fig.suptitle("Residual diagnostics per zone", fontsize=13)

for row, zone in enumerate(ZONES):
    r     = results_dict[zone]
    resid = r["fitted"].resid

    # Residuals over time
    axes[row, 0].plot(resid.index[-500:], resid.values[-500:], lw=0.7, color=r["color"])
    axes[row, 0].axhline(0, color="black", lw=0.8, ls="--")
    axes[row, 0].set_title(f"{r['label']} – Residuals (last 500 obs)")
    axes[row, 0].grid(True, alpha=0.3)

    # ACF of residuals
    plot_acf(resid.dropna(), lags=50, ax=axes[row, 1], title=f"{r['label']} – ACF of residuals", alpha=0.05)

plt.tight_layout()
plt.savefig("05_residuals.png", dpi=150, bbox_inches="tight")
plt.close()
print("[Saved] 05_residuals.png")


# ─────────────────────────────────────────────
# 9. METRICS SUMMARY
# ─────────────────────────────────────────────

print("\n" + "="*55)
print("FORECAST METRICS SUMMARY (24-hour horizon)")
print("="*55)
metrics_df = pd.DataFrame({
    zone: {
        "MAE":  f"{results_dict[zone]['mae']:,.2f}",
        "RMSE": f"{results_dict[zone]['rmse']:,.2f}",
        "MAPE": f"{results_dict[zone]['mape']:.2f}%"
    }
    for zone in ZONES
}).T
metrics_df.index = ZONE_LABELS
print(metrics_df.to_string())


# ─────────────────────────────────────────────
# 10. EXPORT FORECASTS TO CSV
# ─────────────────────────────────────────────

forecast_df = pd.DataFrame({
    zone: results_dict[zone]["forecast"] for zone in ZONES
})
forecast_df.columns = [f"Forecast_{z}" for z in ZONES]
forecast_df.index.name = "Datetime"
forecast_df.to_csv("forecast_24h.csv")
print("\n[Saved] forecast_24h.csv  →  shape:", forecast_df.shape)
print("Done.")
```
